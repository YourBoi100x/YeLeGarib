```python
import numpy as np
import pandas as pd

data = pd.read_csv('2.csv')

concepts = np.array(data.iloc[:, :-1])
target = np.array(data.iloc[:, -1])

def hypothesis_learning(concepts, target):
    specific_h = concepts[0].copy()
    print("Initial specific hypothesis:\n", specific_h)
    
    general_h = [["?" for _ in range(len(specific_h))] for _ in range(len(specific_h))]
    print("Initial general hypothesis:\n", general_h)

    for i, instance in enumerate(concepts):
        if target[i] == "yes":
            print("Positive instance encountered")
            for j in range(len(specific_h)):
                if instance[j] != specific_h[j]:
                    specific_h[j] = '?'
                    general_h[j][j] = '?'
        
        if target[i] == "no":
            print("Negative instance encountered")
            for j in range(len(specific_h)):
                if instance[j] != specific_h[j]:
                    general_h[j][j] = specific_h[j]
                else:
                    general_h[j][j] = '?'
        
        print(f"Step {i + 1}")
        print("Specific hypothesis:", specific_h)
        print("General hypothesis:", general_h)
        print("\n")

    general_h = [h for h in general_h if h != ['?', '?', '?', '?', '?', '?']]
    
    return specific_h, general_h

s_final, g_final = hypothesis_learning(concepts, target)

print("Final Specific Hypothesis:", s_final)
print("Final General Hypothesis:", g_final)
```
